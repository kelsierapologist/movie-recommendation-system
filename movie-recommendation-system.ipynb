```python
import numpy as np
import pandas as pd
import difflib # to get the closest match of the movie name given by the user
from sklearn.feature_extraction.text import TfidfVectorizer  # used to convert textual data into feature vectors
from sklearn.metrics.pairwise import cosine_similarity # to find similarity score
```

```python
# data collection and pre-processing
movies_data = pd.read_csv('/content/movies dataset.csv')
```

```python
movies_data.head()
```

```python
movies_data.shape
```

```python
# feature selection
selected_features = ['genres', 'keywords', 'title', 'cast', 'director']
```

```python
# replacing the missing values with null string
for feature in selected_features:
  movies_data[feature] = movies_data[feature].fillna('')
```

```python
# combining all the five selected features
combined_features = movies_data['genres']+' '+movies_data['keywords']+' '+movies_data['title']+' '+movies_data['cast']+' '+movies_data['director']
```

```python
# converting the text data to feature data
vectorizer = TfidfVectorizer()
feature_vectors = vectorizer.fit_transform(combined_features)
print(feature_vectors)
```

```python
# getting the similarity score using cosine similarity
similarity = cosine_similarity(feature_vectors)
print(similarity)
```

```python
print(similarity.shape)
```

```python
# creating a list with all the movie names in the dataset
movie_list = movies_data['title'].tolist()
print(movie_list)
```

```python
# getting the movie name from the user
movie_name = input('Enter your favourite movie name: ')
```

```python
# finding the close match for the entered movie name
find_close_match = difflib.get_close_matches(movie_name, movie_list)
print(find_close_match)
```

```python
# taking the most close movie name
close_match = find_close_match[0]
print(close_match)
```

```python
# finding the index of the close_match
movie_index = movies_data[movies_data.title == close_match]['index'].values[0]
print(movie_index)
```

```python
# getting the list of similar movies
similarity_score = list(enumerate(similarity[movie_index]))
print(similarity_score)
```

```python
# sorting the movies based on their similarity score
sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True)
print(sorted_similar_movies)
```

```python
# printing the similar movies
print('Movies similar to '+close_match+' are: \n')

i=1
for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = movies_data[movies_data.index==index]['title'].values[0]
  if (i<=10):
    if(title_from_index == close_match):
      i=1
      continue
    print(i, '.', title_from_index)
    i+=1
```

Movie Recommendation System
